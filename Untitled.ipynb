```python
# Importing Dataset
import pandas as pd
import numpy as np
df = pd.read_csv("BKB_WaterQualityData_2020084.csv")
df.head(5)
```

```python
#clean the data so that the range of years is restricted to the 20 year period from 1999 to 2018
df_new = df.query("Year >= 1999 and Year <= 2018")
df_new.head(10)
```

```python
#store the entire dataset back into a new CSV 
df_new.to_csv('back_bay_1998-2018_clean.csv')
```

```python
#importing the new dataset (1999-2018)
DataFrame = pd.read_csv("back_bay_1998-2018_clean.csv")
DataFrame.head(3)
```

```python
#How many and what are the names of the columns in this dataset?
print(len(DataFrame.columns))
print(DataFrame.columns)
```

```python
#What is the mean Dissolved Oxygen (mg/L) over the entire dataset?
print(DataFrame["Dissolved Oxygen (mg/L)"].mean())
```
